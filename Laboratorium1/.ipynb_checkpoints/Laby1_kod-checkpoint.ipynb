```python
import cv2
from matplotlib import pyplot as plt
```

```python
def cv2TOplt (image):
    return_image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    return return_image
```

```python
imgpath = "baboon.png"
```

```python
img = cv2.imread(imgpath)
```

```python
grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
gausBlur = cv2.GaussianBlur(grey, (9,9), 0)
```

```python
while True:
    cv2.imshow("pawian", img)
    cv2.imshow("szary pawian", grey)
    cv2.imshow("rozmyty szary pawian", gausBlur)
    if cv2.waitKey() == 27:
        break
cv2.destroyAllWindows()
```

```python
plt.imshow(cv2TOplt(img))
```

```python
plt.imshow(cv2TOplt(grey))
```

```python
plt.imshow(cv2TOplt(gausBlur))
```
